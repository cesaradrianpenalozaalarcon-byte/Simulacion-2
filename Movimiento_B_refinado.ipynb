```python
import numpy as np
import matplotlib.pyplot as plt
from statistics import mean
from scipy.stats import norm
from math import sqrt
```

Movimiento Browniano

```python
L=500
T=1
dt= T/L
W0= 0


Wi = W0
W_lista = []
t = 0
t1 = []

for i in range (L-1):
  normal = norm.rvs(loc=0,scale=1,size=500) #Normal y cuantas queremos

  Wi = Wi + sqrt(dt) * normal #El wi+1 es igual al anterior mas una normal*raiz(dt)
  W_lista.append(Wi) #Lo guardamos para graficar
  t = t + dt #El tiempo entre los cambios del browniano
  t1.append(t) #Guardamos para graficar

plt.plot(t1,W_lista)
#"y=sqrt(x)"
```
