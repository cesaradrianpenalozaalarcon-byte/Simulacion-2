```python
lamda=4
mu=6
Num_Pkts=10000
count=0
t=0
N=0
tll=expovariate(lamda)

ts=infinity
t_evento_anterior=0
tll_Data=[]
ts_Data=[]
r_Data=[]

Data={}
Área=[]

while count<Num_Pkts:
  if tll<ts:
     t=tll
     tll_Data.append(t)

     Área.append((t-t_evento_anterior)*N)
     t_evento_anterior=t
     N=N+1
     tll= t+expovariate(lamda)
     if N==1:
         ts=t+expovariate (mu)
  else:
     t=ts
     ts_Data.append(t)
     Área.append((t-t_evento_anterior)*N)
     t_evento_anterior=t
     N=N-1
     count=count+1
     if N>0:
         ts= t+expovariate (mu)
     else:
         ts=infinity

for i in range (Num_Pkts):
    d= ts_Data[i]-tll_Data[i]
    r_Data.append(d)

print("El retraso promedio es:W:",round(mean(r_Data),4))
print("El L:",round(sum(Área)/t,4))

#Estimacion de la probabilidad de P(N=k)
for (key,value) in Data.items():
  Data[key]= value/t
#Se verifica que la probabilidad total es  1
print("Suma de probabilidades=", sum(Data.values()))

#Se verifican los valores de expectacion
media=0
for (key,value) in Data.items(): #Esta leyendo el diccionario, por dos columnas key,value, para cada par ordnado suma con la media el producto
  media= media+key*value  #Value es la probabilidad
  print(key,round(value,4)) #Key es un entero
print("Media=",media)
```
